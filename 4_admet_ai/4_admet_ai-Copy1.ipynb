```python
def split(df, chunk_size):
    num_chunks = len(df) // chunk_size
    dataframes = {}

    if len(df) % chunk_size != 0 or len(df) == 0:
        num_chunks += 1
    for i in range(num_chunks):
        chunk = df[i * chunk_size:(i + 1) * chunk_size].copy()
        dataframes[f'df_{i + 1}'] = chunk

    return dataframes
```

```python
#New dataframe with the remaining rows of the data set
data2 = data[50000:]
```

```python
#Split of the data set due to limited computational resources
chunked_dfs = split(data2, 25000)
```

```python
chunked_dfs 
```

```python
#Function to calculate and save results for each split dataframe 
def ADMET(dfs, SMILES_column='Canonical_SMILES_STD', output_dir='output'):

    import os
    os.makedirs(output_dir, exist_ok=True)

    for name, chunk_df in dfs.items():
        # Get the SMILES strings from the specified column
        smiles_list = chunk_df[SMILES_column].tolist()

        # Make predictions using the model
        pred = model.predict(smiles=smiles_list)

        # Ensure predictions are in a DataFrame format
        pred_df = pd.DataFrame(pred)  # Convert predictions to DataFrame

        # Add the SMILES column to the predictions DataFrame for merging
        pred_df[SMILES_column] = smiles_list

        # Merge the original DataFrame with predictions based on the SMILES column
        results_df = pd.merge(chunk_df, pred_df, on=SMILES_column, how='left')

        # Save the results to a CSV file
        csv_filename = f"{output_dir}/{name}_predictions.csv"  # File name based on chunk name
        results_df.to_csv(csv_filename, index=False)  # Save as CSV

        print(f"Saved predictions for {name}: {csv_filename}")  # Confirmation message
```

```python
#Calculate ADMET descriptors with previous function
ADMET(chunked_dfs)
```
